```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url=['https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc&ref_=adv_prv','https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc&start=51&ref_=adv_nxt','https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc&start=101&ref_=adv_nxt','https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc&start=151&ref_=adv_nxt','https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc&start=201&ref_=adv_nxt']
```

```python
page=[]
soup=[]

for i in url:
    page.append(requests.get(i))

for i in page:
    soup.append(BeautifulSoup(i.text,'html.parser'))
```

# Movie Id

```python
element=[]
movie_id=[]
for i in soup:
    element.append(i.find_all(class_='lister-item-index unbold text-primary'))
for item in element:
    for i in item:
        movie_id.append(i.text)
```

# Title

```python
title_element=[]
title=[]
for i in soup:
    title_element.append(i.find_all('h3',{'class':'lister-item-header'}))
for item in title_element:
    for i in item:
        title.append(i.find('a').text)
```

# About

```python
description_element=[]
description=[]
about=[]
for i in soup:
    description_element.append(i.find_all('p',{'class':'text-muted'}))
for item in description_element:
    for i in item:
        description.append(i.text)
for i in range(len(description)):
    if(int(i)%2!=0):
        about.append(description[i])
for i,j in enumerate(about):
    about[i]=j.replace('\n','').strip()
```
